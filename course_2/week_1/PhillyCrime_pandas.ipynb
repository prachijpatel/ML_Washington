```python
import pandas as pd
from matplotlib import pyplot as plt
sales=pd.read_csv("../data/Philadelphia_Crime_Rate_noNA.csv")
sales
```

```python
%matplotlib notebook
plt.scatter(sales["CrimeRate"], sales["HousePrice"])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
lr = LinearRegression()
X=sales["CrimeRate"]
X = X[:, np.newaxis]
model=lr.fit(X,sales["HousePrice"],sample_weight=None)
predications=lr.predict(X)
predications
```
